Q1) Scatter Plot (sns.scatterplot()):

Uses: Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, trends, and correlations in the data. Scatter plots are useful for assessing the strength and direction of the relationship between variables.
Histogram (sns.histplot()):

Uses: Histograms are used to visualize the distribution of a single continuous variable. They display the frequency or proportion of data points falling within different intervals (bins) along the x-axis. Histograms are useful for understanding the central tendency, spread, and shape of the data distribution.
Bar Plot (sns.barplot()):

Uses: Bar plots are used to visualize the distribution of a categorical variable or the relationship between a categorical variable and a continuous variable. They display the mean or aggregate value of the dependent variable for each category of the independent variable. Bar plots are useful for comparing the values of different categories or groups.
Box Plot (sns.boxplot()):

Uses: Box plots, also known as box-and-whisker plots, are used to visualize the distribution of a continuous variable and identify potential outliers. They display the five-number summary (minimum, first quartile, median, third quartile, and maximum) of the data distribution. Box plots are useful for comparing the spread and central tendency of multiple distributions.
Heatmap (sns.heatmap()):

Uses: Heatmaps are used to visualize the relationships between two categorical variables or between one categorical and one continuous variable. They display the values of a matrix as colors, where each cell represents the relationship between a pair of variables. Heatmaps are useful for identifying patterns and correlations in large datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plotting line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plotting box plots
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plotting histogram with hue for 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plotting pair plot with hue for 'species' column
sns.pairplot(iris_data, hue='species')

# Adjusting layout
plt.subplots_adjust(top=0.95)
plt.suptitle('Pair Plot of Iris Dataset by Species', fontsize=16)

plt.show()
```
